<h3>DIVIDING THE INITIAL DATASET INTO 6 CSV FILES </h3>

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# 1. Загрузка данных
# Убедись, что файл называется source_data.csv или поменяй название тут
df = pd.read_csv('source_data.csv')

print("Исходные данные загружены. Строк:", len(df))

# --- ЭТАП 1: Создание справочников (Dimensions) ---

# 1. Таблица product_importance
# Извлекаем уникальные значения важности
importance_df = df[['Product_importance']].drop_duplicates().reset_index(drop=True)
importance_df['importance_id'] = importance_df.index + 1
importance_df = importance_df.rename(columns={'Product_importance': 'importance_name'})
# Сохраняем
importance_df.to_csv('product_importance.csv', index=False)
print("-> Создан файл product_importance.csv")

# 2. Таблица shipment_mode
mode_df = df[['Mode_of_Shipment']].drop_duplicates().reset_index(drop=True)
mode_df['mode_id'] = mode_df.index + 1
mode_df = mode_df.rename(columns={'Mode_of_Shipment': 'mode_name'})
# Сохраняем
mode_df.to_csv('shipment_mode.csv', index=False)
print("-> Создан файл shipment_mode.csv")

# 3. Таблица warehouse
warehouse_df = df[['Warehouse_block']].drop_duplicates().reset_index(drop=True)
warehouse_df['warehouse_id'] = warehouse_df.index + 1
warehouse_df = warehouse_df.rename(columns={'Warehouse_block': 'warehouse_block'})
# Сохраняем
warehouse_df.to_csv('warehouse.csv', index=False)
print("-> Создан файл warehouse.csv")


# --- ЭТАП 2: Создание основных сущностей ---

# 4. Таблица products
# Нам нужно создать уникальные продукты. Предположим, что уникальная комбинация 
# Цены, Веса и Важности - это и есть уникальный товар.
products_cols = ['Cost_of_the_Product', 'Weight_in_gms', 'Product_importance']
products_df = df[products_cols].drop_duplicates().reset_index(drop=True)
products_df['product_id'] = products_df.index + 1

# Присоединяем importance_id (заменяем текст на число)
products_df = products_df.merge(importance_df, left_on='Product_importance', right_on='importance_name')

# Формируем финальную таблицу продуктов
final_products = pd.DataFrame()
final_products['product_id'] = products_df['product_id']
final_products['cost'] = products_df['Cost_of_the_Product']
# Придумаем базовую цену чуть выше стоимости, чтобы было реалистично
final_products['base_price'] = final_products['cost'] * 1.2 
final_products['weight_gms'] = products_df['Weight_in_gms']
final_products['importance_id'] = products_df['importance_id']

final_products.to_csv('products.csv', index=False)
print("-> Создан файл products.csv")

# 5. Таблица customers
# В этом датасете нет имен, поэтому мы создадим клиента для каждой транзакции,
# или сгруппируем уникальных. Для простоты задачи ER-диаграммы: 
# будем считать ID из файла = customer_id (1 клиент = 1 заказ в этом срезе)
customers_df = df[['ID', 'Gender', 'Prior_purchases']].copy()
customers_df = customers_df.rename(columns={'ID': 'customer_id', 'Gender': 'gender', 'Prior_purchases': 'prior_purchases'})
customers_df.to_csv('customers.csv', index=False)
print("-> Создан файл customers.csv")


# --- ЭТАП 3: Главная таблица (Fact Table) ---

# 6. Таблица shipments
# Берем исходную таблицу и заменяем все текстовые поля на ID, которые мы создали выше

shipments_df = df.copy()

# Подтягиваем mode_id
shipments_df = shipments_df.merge(mode_df, left_on='Mode_of_Shipment', right_on='mode_name')
# Подтягиваем warehouse_id
shipments_df = shipments_df.merge(warehouse_df, left_on='Warehouse_block', right_on='warehouse_block')
# Подтягиваем product_id (связываем по цене, весу и важности)
shipments_df = shipments_df.merge(products_df, on=['Cost_of_the_Product', 'Weight_in_gms', 'Product_importance'])

# Генерируем случайные даты для shipment_date (так как их нет в исходнике)
def random_date(start, end):
    return start + timedelta(days=random.randint(0, (end - start).days))

start_date = datetime(2023, 1, 1)
end_date = datetime(2024, 1, 1)
shipments_df['shipment_date'] = [random_date(start_date, end_date) for _ in range(len(shipments_df))]
```

```python
# Формируем финальную структуру как в SQL
final_shipments = pd.DataFrame()
final_shipments['shipment_id'] = shipments_df['ID'] # ID из файла
final_shipments['customer_id'] = shipments_df['ID'] # Тот же ID, так как мы решили 1 к 1
final_shipments['product_id'] = shipments_df['product_id']
final_shipments['warehouse_id'] = shipments_df['warehouse_id']
final_shipments['mode_id'] = shipments_df['mode_id']
final_shipments['shipment_date'] = shipments_df['shipment_date']
final_shipments['reached_on_time'] = shipments_df['Reached_on_time_Y_N']
final_shipments['customer_care_calls'] = shipments_df['Customer_care_calls']
final_shipments['customer_rating'] = shipments_df['Customer_rating']
final_shipments['discount_offered'] = shipments_df['Discount_offered']

final_shipments.to_csv('shipments.csv', index=False)
print("-> Создан файл shipments.csv")
print("\nГотово! Теперь у вас есть 6 файлов для загрузки в базу данных.")
```
